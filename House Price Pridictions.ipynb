```python

x = df1.drop(columns=['price','yr_renovated','zipcode'],axis=1)
y = df1['price']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
#X_train=ms.fit_transform(x_train)
#X_test=ms.transform(x_test)
```

```python
import ubml
metrics_performance,best_model=ubml.model_train_test(mode='regression',x_train=x_train,y_train=y_train,x_test=x_test,y_test=y_test)

```

```python
print(best_model)
metrics_performance
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR=RandomForestRegressor(max_depth=25,n_estimators=100,random_state=42)
RFR.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error

pred_train=RFR.predict(x_train)
pred_test=RFR.predict(x_test)

train_error=mean_squared_error(y_train,pred_train)
test_error=mean_squared_error(y_test,pred_test)

print("training error :",train_error)
print('test error :',test_error)
```

```python
ypred=RFR.predict(x_test)
score=RFR.score(x_test,y_test)
print('accuracy : ',score)
```

```python
plt.figure(figsize=(5,5))
ax=sns.distplot(y_test,hist=False,color='r',label='Actual value')
sns.distplot(ypred,hist=False,color='b',label='predicted value',ax=ax)
plt.title('Actual value and predicited value')
plt.show()
```
